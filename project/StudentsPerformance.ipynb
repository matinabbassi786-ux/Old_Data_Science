```python
df = pd.read_csv("StudentsPerformance.csv")

labels=["female","male"]
plt.pie([df.query("gender == 'female'").count()["gender"],df.query("gender == 'male'").count()["gender"]] ,colors=["pink","blue"],startangle=90,labels=labels,autopct='%1.1f%%')
plt.title("How many gender are thera")
```

```python
group_count = [ 
df[df["race&ethnicity"] =="group A"].count()["gender"],
df[df["race&ethnicity"] =="group B"].count()["gender"],
df[df["race&ethnicity"] =="group C"].count()["gender"],
df[df["race&ethnicity"] =="group D"].count()["gender"],
df[df["race&ethnicity"] =="group E"].count()["gender"],
]
labels =[ "group A" ,"group B","group C","group D","group E" ]
color = [ "#F0F8FF" , "#F5F5DC" , "#5F9EA0" , "#F0E130" ,"#C2B280"]
plt.pie(group_count,labels=labels, startangle=360,autopct='%1.1f%%', colors=color)
plt.title(" Is Show How Many Students in group A to group E")
```

```python
group_count = [
df[df["parental level of education"] == "bachelor's degree"].count()["lunch"],
df[df["parental level of education"] == "master's degree"].count()["lunch"],
df[df["parental level of education"] == "some college"].count()["lunch"],
df[df["parental level of education"] == "associate's degree"].count()["lunch"],
df[df["parental level of education"] == "high school"].count()["lunch"],
df[df["parental level of education"] == "some high school"].count()["lunch"],
]
labs = ["bachelor's degree", "master's degree" ,"some college","associate's degree" , "high school","some high school" ]
plt.pie(group_count,labels=labs,autopct='%1.1f%%')
plt.title("This Pie Chart Show How Many Students hava parental level of education")
```
