```python
def Shuffle_and_split(data,teast_ratio):
    np.random.seed(42)
    Shuffle_index = np.random.permutation(len(data))
    test_set_sizer = int(len(data)*teast_ratio)
    test_index = Shuffle_index[:test_set_sizer]
    TRAIN_index =Shuffle_index[test_set_sizer:]
    return data.iloc[TRAIN_index],data.iloc[test_index]
```

```python
treain , teast = Shuffle_and_split(df,0.4)
```

```python
df = df.sort_values(by="ocean_proximity",ascending=False ).dropna().reset_index()
```

```python
df["ocean_proximity"].unique()
```

```python
NEAR_OCEAN = df[df["ocean_proximity"] == "NEAR OCEAN"].sample(n=int((20/100)*len(df[df["ocean_proximity"] == "NEAR OCEAN"]))).copy()
NEAR_BAY =  df[df["ocean_proximity"] == "NEAR BAY"].sample(n=int((20/100)*len(df[df["ocean_proximity"] == "NEAR BAY"]))).copy()
ISLAND =  df[df["ocean_proximity"] == "ISLAND"].sample(n=int((100/100)*len(df[df["ocean_proximity"] == "ISLAND"]))).copy()
INLAND = df[df["ocean_proximity"] == "INLAND"].sample(n=int((20/100)*len(df[df["ocean_proximity"] == "INLAND"]))).copy()
OCEAN = df[df["ocean_proximity"] == "<1H OCEAN"].sample(n=int((20/100)*len(df[df["ocean_proximity"] == "<1H OCEAN"]))).copy()
```

```python
NEAR_BAY = NEAR_BAY.drop('index',axis=1)
NEAR_OCEAN = NEAR_OCEAN.drop('index',axis=1)
ISLAND = ISLAND.drop('index',axis=1)
INLAND = INLAND.drop('index',axis=1)
OCEAN = OCEAN.drop('index',axis=1)
```

```python
data = {
    "longitude": np.array(NEAR_BAY["longitude"],NEAR_OCEAN["longitude"],ISLAND["longitude"],INLAND["longitude"],OCEAN["longitude"])
    "latitude": np.array(NEAR_BAY["latitude"],NEAR_OCEAN["latitude"],ISLAND["latitude"],INLAND["latitude"],OCEAN["latitude"])
    "housing_median_age": np.array(NEAR_BAY["housing_median_age"],NEAR_OCEAN["housing_median_age"],ISLAND["housing_median_age"],INLAND["housing_median_age"],OCEAN["housing_median_age"])
    "total_rooms":  np.array(NEAR_BAY["total_rooms"],NEAR_OCEAN["total_rooms"],ISLAND["total_rooms"],INLAND["total_rooms"],OCEAN["total_rooms"])
    "total_bedrooms":  np.array(NEAR_BAY["total_bedrooms"],NEAR_OCEAN["total_bedrooms"],ISLAND["total_bedrooms"],INLAND["total_bedrooms"],OCEAN["total_bedrooms"])
    "population": np.array(NEAR_BAY["population"],NEAR_OCEAN["population"],ISLAND["population"],INLAND["population"],OCEAN["population"])
    "households": np.array(NEAR_BAY["households"],NEAR_OCEAN["households"],ISLAND["households"],INLAND["households"],OCEAN["households"])
    "median_income": np.array(NEAR_BAY["median_income"],NEAR_OCEAN["median_income"],ISLAND["median_income"],INLAND["median_income"],OCEAN["median_income"])
    "median_house_value": np.array(NEAR_BAY["median_house_value"],NEAR_OCEAN["median_house_value"],ISLAND["median_house_value"],INLAND["median_house_value"],OCEAN["median_house_value"])
    "ocean_proximity": np.array(NEAR_BAY["ocean_proximity"],NEAR_OCEAN["ocean_proximity"],ISLAND["ocean_proximity"],INLAND["ocean_proximity"],OCEAN["ocean_proximity"])
}
```
